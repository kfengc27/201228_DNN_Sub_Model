```python
import ipywidgets as widgets
import matplotlib
import matplotlib.pyplot as plt
from tensorflow import keras 
import pandas as pd 
from IPython.display import display, clear_output
import time
import tensorflow as tf 
import numpy as np
pd.options.mode.chained_assignment = None  # default='warn'  remove warn 
from IPython.display import clear_output
import time 
from statistics import mean 
import statistics 
from statistics import mode 
```

```python
# load dnn model
load_model = keras.models.load_model('dnn_model')
sub_model = tf.keras.Model(load_model.get_layer(index = 0).output, load_model.get_layer(index = 2).output)

# read the nature of each question
fb_content = pd.read_csv('./Input/fb_content.csv')

def Average(lst): 
    return mean(lst) 

def most_common(List): 
    return(mode(List)) 
```

```python
column_name = ["Question_Number", "factor_be_influenced_max", "factor_be_influenced_max_most_common", "factor_be_influenced_min", "factor_be_influenced_min_most_common"]
factor_results = pd.DataFrame(columns = column_name) 
# There are total 597 participants 
for n in range(1, 2):
    column_name = 'Q1.9_' + str(n)
    print(column_name)
    factor_influence_max_list = []
    factor_influence_min_list = []
    for k in range(0,1):
        base_input_content = fb_content.iloc[k:k+1]
        base_input = sub_model.predict(base_input_content)
        sub_layer = tf.keras.layers.Softmax()
        base_result = sub_layer(base_input).numpy()
        print("Base_Result Input with %s"%(str(fb_content.iloc[k:k+1][column_name][0])))
        print(base_result)
        print('\n')
        input_content = base_input_content
        for i in range(1,8):
            #TODO: loop column 
            input_content[column_name] = i
            inp = sub_model.predict(input_content)
            sub_layer = tf.keras.layers.Softmax()
            result = sub_layer(inp).numpy()
            difference = result-base_result
            max_index = difference[0].tolist().index(max(difference[0]))
            min_index = difference[0].tolist().index(min(difference[0]))
            max_number = float(max(difference[0]))
            min_number = float(min(difference[0]))
            mean_number = Average(difference[0])
            factor_influence_max_list.append(max_index)
            factor_influence_min_list.append(min_index)
#     factor_list = [column_name, max(factor_influence_max_list),most_common((factor_influence_max_list)),min(factor_influence_min_list), most_common(factor_influence_min_list)]
#     factor_result_series = pd.Series(factor_list, index = factor_results.columns)
#     print(factor_result_series)
#     factor_results = factor_results.append(factor_result_series, ignore_index = True)
# today = time.strftime("%Y-%m-%d")
# file_name = today + '_factor_results' + '.csv'
# factor_results.to_csv(file_name)
            print("Input %d:"%i)
            print(result[0])
            print("Diff: ")
            print(difference[0])
            print("\nMax Changed Factor %s || Min Changed Factor %s"%(max_index, min_index))
            print("Max: %.3f Min: %.3f Mean: %.3f\n"%(max_number, min_number, mean_number))
    print("factor_be_influenced_max", max(factor_influence_max_list))
    print("factor_be_influenced_max_most_common", most_common((factor_influence_max_list)))
    print("factor_be_influenced_min", min(factor_influence_min_list))
    print("factor_be_influenced_min_most_common", most_common(factor_influence_min_list))
#         time.sleep(10)
#         clear_output()
```
